```python
# 파이썬 설치 주의사항
- Add Python 버전 to PATH 체크 -> 명령 프롬프트 어디에서나 파이썬 실행(어느 폴더에 있더라도 실행 가능)
- Disable path length limit -> 기본적으로 Windows는 파일 이름과 경로의 길이를 260자로 제한, 260자를 넘더라도 문제가 발생하지 않도록 길이 제한을 해제

# IDLE 실행
- 파이썬 정상 설치를 확인하기 위해 실행(>>> 뒤에 커서가 깜박이면 정상)
- IDLE(Integrated Development and Learning Environment) : 통합 개발 학습 환경

# 프로젝트 폴더 생성
- 소스 파일을 저장할 폴더 C:\project를 생성(로컬 디스크 C에 생성)

# 리눅스에 파이썬 설치하기
- 우분투 리눅스 최신 버전은 파이썬 3가 미리 설치되어 있음(우분투는 배포판)
파이썬을 실행하려면 $ python3 입력
- CentOS는 파이썬 3가 아닌 2.7이 설치되어 있을 수 있음
파이썬 3를 설치하려면
$ sudo yum install -y epel-release
$ sudo yum install -y python3* 
- 리눅스 배포판의 패키지 대신 파이썬 소스 코드를 직접 컴파일하여 설치
- 먼저 C 컴파일러와 빌드 도구를 설치
(우분투에서 설치하는 방법)
$ sudo apt-get update
$ sudo apt-get install build-essential
(CentOS에서 설치하는 방법)
$ sudo yum groupinstall 'Development Tools'
- 파이썬 소스 코드를 받아서 컴파일한 뒤 설치
$ curl -0 https://www.python.org/ftp/python/3.6.0/Python-3.6.0.tgz
$ tar vxzf Python-3.6.0.tgz
$ cd Python-3.6.0
$ /configure
$ make
$ sudo make install
- 리눅스에서 IDLE을 사용하려면 Tklnter를 설치
(우분투에서 설치하는 방법)
$ sudo apt-get install python3-tk
(CentOS에서 설치하는 방법)
$ sudo yum install -y epel-release
$ sudo yum install python3*-tkinter
- macOS에 파이썬 설치하기
Max OS X 64-bit/32-bit installer 설치(파이썬 다운로드 사이트에서)
```

```python
# 인터프리터(interpreter) 방식
코드를 한 줄 한 줄 실행하여 결과를 얻는 방식

# 파이썬 셸(Python shell)
- 파이썬 코드를 길게 직접 입력해서 실행하는 프로그램
- 파이썬 셸은 파이썬 인터프리터와 대화하듯이 코드를 처리한다고 해서 대화형 셸 또는 인터렉티브 모드(interactive mode)라고도 함
- REPL(Read-Eval-Print Loop) 방식 : 코드를 읽고, 평가(계산, 실행), 출력한다.

# 파이썬 프롬프트(Python prompt)
>>> 부분

# IDLE의 파이썬 셸에서 에러가 났을 때
- IDLE의 파이썬 셸에서는 ↑ 방향 키를 누르면 이전 코드로 쉽게 돌아갈 수 있음
- 이전 코드에서 엔터 키를 누르면 해당 코드를 다시 사용할 수 있음
(동일한 코드가 복사되어 나타나고, 그것을 올바른 코드로 수정하여 엔터 키 누르기)
```

```python
# IDLE에서 소스 파일 실행하기
- 소스 코드 또는 소프 파일 : 프로그래밍 코드를 소스 파일로 저장한 것
- IDLE은 파이썬 셸을 통해 코드 실습, 결과 확인, 소스 파일 편집, 실행 기능을 함께 제공
- IDLE을 실행하고 메뉴에서 File > New File을 클릭
- 소스 코드 입력
- 입력한 파이썬 코드를 실행하려면 .py 파일에 저장(C:\project 폴더에 저장), 소스 코드 편집 창의 메뉴에서 File > Save를 클릭하거나 Ctrl+S를 누름
- 소스 파일 실행 Run > Run Module 또는 F5, 파이썬 셸 창에 결과 출력
- 파이썬 코드를 저장한 .py 파일을 파이썬 스크립트(Python script)라고 부름
```

```python
# 명령 프롬프트에서 파이썬 사용하기
- Win+R을 누른 뒤 cmd를 입력하여 명령 프롬프트를 실행 ★유용하네!!
- python을 입력하여 파이썬 셸을 실행
- print("Hello, world!")를 입력하고 엔터 키
- 명령 프롬프트에서 파이썬 셸을 끝내려면 exit()을 입력하거나 Ctrl+Z를 누름

# 에러(명령 프롬프트에서 파이썬이 실행되지 않을 때
- 파이썬 삭제, 다시 설치
- 제어판 > 프로그램 > 프로그램 제거 > Python 3.9.0 (64-bit)를 클릭
- Uninstall 클릭
- 파이썬 다시 설치하면서 Add Python 3.9 to PATH에 체크
- 제어판 > 시스템 및 보안 > 시스템 > 고급 시스템 설정 > 환경 변수(N)...에서 <사용자이름>에 대한 사용자 변수(U)에서 Path를 선택하고 편집(E)..를 클릭
- 새로 만들기(N)를 클릭한 뒤 다음 두 경로를 추가(설치 버전과 비트수에 맞게)
32비트일 경우
C:\Users\<사용자이름>\AppData\Local\Programs\Python\Python36-32\ 또는
C:\Users\<사용자이름>\AppData\Local\Programs\Python\Python36-32\Scripts\
64비트일 경우
C:\Users\<사용자이름>\AppData\Local\Programs\Python\Python36\ 또는
C:\Users\<사용자이름>\AppData\Local\Programs\Python\Python36\Scripts\

# 리눅스와 macOS에서 파이썬 사용하기
- 리눅스와 macOS에서는 보통 콘솔(터미널)에서 python3를 사용
$ python3 입력
- 리눅스, macOS 콘솔에서 파이썬 셸을 끝내려면 exit() 또는 Ctrl+D

# 콘솔의 파이썬 셸에서 이전 코드 사용하기
입력한 코드에서 에러가 발생했다면 ↑ 방향키와 엔터키를 눌러 에러가 발생한 코드를 찾은 뒤에 올바르게 수정해서 실행 

# 명령 프롬프트에서 스크립트 파일 실행하기
- 명령 프롬프트에서 hello.py 파일을 실행
- Win+R을 누른 뒤 cmd를 입력하여 명령 프롬프트를 실행
- C:\project 폴더로 이동(cd는 change directory의 줄인말)
C:\Users\최예름>cd c:\project
- python hello.py를 입력하여 스크립트 파일을 실행(python 스크립트파일.py 형식)
c:\project>python hello.py

# 리눅스와 macOS에서 스크립트 파일 실행하기
- 리눅스와 macOS에서는 python3를 이용해서 스크립트 파일을 실행(python3 스크립트파일.py 형식)
$ python hello.py
Hello, world!
- 리눅스와 macOS에서는 앞에 python3를 붙이지 않고 .py 파일을 그대로 실행할 수도 있음
- 먼저 다음 내용을 hello.py로 저장(리눅스 콘솔에서 파일을 저장하려면 nano, vim, emacs 등 텍스트 편집기 사용
#!/usr/bin/python3
print('Hello, world!')

# 리눅스와 macOS에서 스크립트 파일 실행하기
- hello.py 파일을 저장했으면 chmod 명령으로 hello.py에 실행 권한을 줌(+x 옵션은 실행 권한을 준다는 뜻이며 chmod 755도 같은 기능). 
r(읽기), w(쓰기), x(실행)
*왜 755인가?
rwx
111(2진수) -> 7(10진수)
rwx
101(2진수) -> 5(10진수)
rwx
rwx
101(2진수) -> 5(10진수)
rwx
- 그다음에 ./hello.py를 입력하여 파이썬 파일을 실행(./는 현재 디렉터리에 있는 파일을 실행한다는 뜻
$ chmod +x hello.py
$ ./hello.py
- hello.py 파일의 첫 줄에 있는 #!/usr/bin/python3가 핵심, 이 부분을 셔뱅(shebang)
- 셔뱅은 현재 파일을 실행해 줄 프로그램을 지정할 때 사용
- 셔뱅에 파이썬 경로를 지정해 놓으면 해당 파이썬으로 .py 파일을 실행
- 파이썬이 설치된 경로는 시스템마다 다를 수 있으므로 보통 다음과 같이 /usr/bin/env를 사용하여 파이썬 경로를 지정함(env는 환경 변수에 지정된 경로에서 프로그램을 찾아서 실행)
#!/usr/bin/env python3
```
