```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
# cyl 값이 4인 컬럼을 출력하려면?
mpg.query('cyl == 4')
```

```python
# hwy가 30 이상인 경우를 추출하려면?
mpg.query('hwy >= 30')
```

```python
# 2000년도부터 생산된 것들 중에서 displ이 2.0 이상인 것을 추출하면?
mpg.query('displ >= 2.0 & year >= 2000')
```

```python
# cyl이 5일 때 컬럼을 추출하면?
cyl1 = mpg.query('cyl == 5')
cyl1
```

```python
# cyl1 으로 출력된 항목에서 cty의 평균을 구하면?
cyl1['cty'].mean()
```

```python
# cyl1 으로 출력된 항목에서 sum을 구하면?
cyl1['cty'].sum()
```

```python
# cyl1에 new 컬럼을 만들고 Sell 값을 넣으려면?
cyl1['new'] = 'Sell'
cyl1
```

```python
# cyl1의 정보를 출력하려면?
cyl1.info()
```

```python
# cty와 hwy만 출력하기
mpg[['cty','hwy']]
```

```python
# 1999년도의 manufacturer를 추출하려면?
mpg.query('year==1999')['manufacturer']
```

```python
# year 오름차순으로 출력하려면?
mpg.sort_values('year')
```

```python
# year 내림차순으로 출력하려면?
mpg.sort_values('year', ascending=False)
```

```python
# manufacturer와 model을 함께 정렬하려면?
mpg.sort_values(['manufacturer','model'])
```

```python
# tot = cty + hwy 를 새로운 컬럼으로 추가
mpg.assign(tot = mpg['cty']+mpg['hwy'])
```

```python
# cty와 hwy의 합인 tot을 lambda식으로 전처리하려면?
mpg.assign(tot = lambda x: x['cty'] + x['hwy'])
```
